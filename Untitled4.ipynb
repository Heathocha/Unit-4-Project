```python
import numpy as np
import scipy as sp
import pandas as pd
import random
import seaborn as sns
import re
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('ign.csv')
df.head()
```

```python
df['genre'].value_counts()
```

```python
Action = ["Action", "Action, RPG", "Action, Adventure", "Action, Compilation", "Other, Action", "Action, Editor", "Action, Strategy", "Action, Puzzle"]
Sports = ["Sports", "Sports, Action", "Wrestling", "Baseball", "Sports, Golf", "Wrestling, Simulation", "Sports, Party", "Sports, Editor", "Sports, Other", "Sports, Fighting"]
Shooter = ["Shooter", "Shooter, Adventure"]                     
Racing = ["Racing", "Racing, Action"]
Adventure = ["Adventure", "Adventure, Platformer" "Adventure, Adult"]
Strategy = ["Strategy", "Strategy, RPG", "Strategy, Compilation", "Strategy, Simulation"]
RPG = ["RPG", "RPG, Action", "Compilation, RPG", "Puzzle, RPG"]                          
Platformer = ["Platformer", ]                   
Puzzle = ["Puzzle", "Puzzle, Action", "Puzzle, Platformer"] 

def mapping(item):
    if item in Action:
        return "Action"
    if item in Sports:
        return "Sports"
    if item in Shooter:
        return "Shooter"
    if item in Racing:
        return "Racing"
    if item in Adventure:
        return "Adventure"
    if item in Strategy:
        return "Strategy"
    if item in RPG:
        return "RPG"
    if item in Platformer:
        return "Platformer"
    if item in Puzzle:
        return "Puzzle"
    return "Other"

df["Type"] = df["genre"].map(mapping)
```
